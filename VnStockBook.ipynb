```python
from vnstock import *


import talib
import pandas as pd

# Chuẩn bị dữ liệu đầu vào
df = stock_historical_data('VND', start_date='2022-10-01', end_date='2023-02-16')
close = df['Close'].values

# Tính toán các chỉ số kỹ thuật
rsi = talib.RSI(close, timeperiod=14)
ma10 = talib.SMA(close, timeperiod=10)
ma20 = talib.SMA(close, timeperiod=20)
ma50 = talib.SMA(close, timeperiod=50)
ma100 = talib.SMA(close, timeperiod=100)
upper, middle, lower = talib.BBANDS(close, timeperiod=20, nbdevup=2, nbdevdn=2)

# Kiểm tra điều kiện xu hướng tăng
print(close[-1])
if close[-1] > ma10[-1] and close[-1] > ma20[-1] and close[-1] > ma50[-1] and close[-1] > ma100[-1] and rsi[-1] > 50 and close[-1] > lower[-1]:
    print("Xu hướng tăng")
else:
    print("Không có xu hướng tăng")

```
